## Case 1: Flow = 15m/hr Tower Pressure = 30Kpag

```python
import sympy as sp
import numpy as np
```

```python
hL, L, D, v, g = sp.symbols(['hL', 'L', 'D', 'v', 'g'])

hL_ = 0.0819*(L/D)*v**2/g  # v [m/s]
```

```python
#Flow, t, Area = sp.symbols(['Flow', 't', 'Area'])
Flow = 15 # Flow rate [m/hr]
t = 60*60 # time (converting hr to seconds) [s]
D = 2*25.4E-3 # pipe diamenter [m]
Area = np.pi*D**2/4 # pipe sectional area [m]
v = Flow/Area/t
v
```

```python
Flow, t, Area = sp.symbols(['Flow', 't', 'Area'])
v2 = Flow /Area /t
v2_ = v2.subs(Flow, 15).subs(t, 3600).subs(Area, np.pi*(2*25.4E-3)**2/4)
```

```python
Flow, t, AreaTank = sp.symbols(['Flow', 't', 'Area'])
v1 = Flow /Area /t
v1_ = v1.subs(Flow, 15).subs(t, 3600).subs(AreaTank, np.pi*(8)**2/4)
```

```python
L = 1 + 13.5 + 1
D = 25.4E-3
L,D,v,g = sp.symbols(['L','D','v','g'])
hL=0.0819*(L/D)*v**2/g
```

```python
hL_ = hL.subs(L, 15.5).subs(v, v2_).subs(D, 2*25.4E-3).subs(g, 9.81)
hL_
```

```python
v1, v2, g, Z1, Z2, P1, P2, rho, hl = sp.symbols(['v1', 'v2', 'g', 'Z1', 'Z2', 'P1', 'P2', 'rho', 'hL']) 
Wp = (Z2-Z1) + (P2-P1)/(rho*g) + (v2**2-v1**2)/(2*g) + hl
Wp
```

```python
Wp.subs(Z1, 0.5).subs(Z2, 13.5).subs(hl, hL_).subs(v1, v1_).subs(v2, v2_).subs(g, 9.81).subs(P1, 101325).subs(P2, 131325).subs(rho, 1000)
```

```python
El_Loss = Z2-Z1
Elevation_Loss = El_Loss.subs(Z2,13.5).subs(Z1,0.5)
Elevation_Loss
```

```python
P_Loss = (P2-P1)/(rho*g)
Pressure_Loss = P_Loss.subs(P2, 131325).subs(P1, 101325).subs(rho, 1000).subs(g, 9.81)
Pressure_Loss
```

```python
Vel_Loss = (v2**2-v1**2)/(2*g)
Velocity_Loss = Vel_Loss.subs(v1, v1_).subs(v2, v2_).subs(g, 9.81)
Velocity_Loss
```

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline


# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['Elevation Head Loss', 'Pressure Head Loss', 'Velocity Head Loss', 'Friction Head Loss']
Head_Losses = [Elevation_Loss, Pressure_Loss, Velocity_Loss, hL_]

fig, ax = plt.subplots()
ax.pie(Head_Losses, labels=labels, autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('Head Losses')

plt.show()
```
